```python
lines = sc.textFile("/FileStore/tables/steve.txt")
```

```python
## Prints first paragraph
text_file = sc.textFile("/FileStore/tables/steve.txt")
print(lines.take(5))
```

```python
# Split the text into list of words
words = text_file.flatMap(lambda line: line.split(" "))
list = words.collect() #found the number of words to print all words in the text
size = len(list)
print(len(list))
for a in words.take(size):
  print(a)
```

```python
#List of key value pairs, where the word is the key and 2 is the value
pairs = words.map(lambda s: (s,1)) #changed the value, so that each word weighs 1
```

```python
counts = pairs.reduceByKey(lambda a,b: a+b)
listCounts = counts.collect() #as the take function takes into arguments the number of n first elements, hasd to find the size of the list
sizeC = len(listCounts)
print(sizeC)
for (count, word) in counts.take(sizeC):
  print ("%s: %i"%(count, word))
  
#What are we exactly counting ?
#We are exactly counting the number of occurences of a word, times its value. In this case, "out:" appears three times in the text, whereas "more."[12]" appears only once
```

```python
edges_text = sc.textFile("/FileStore/tables/edgelist.txt")
idsLabels =  sc.textFile("/FileStore/tables/idslabels.txt")
lines = edges_text.flatMap(lambda line: line.split("\n"))
size = len(lines.collect())

text = edges_text.flatMap(lambda line: line.split(" ")[1:])
#creates couples of id and line
pairs = text.map(lambda s: (s,1))
counts = pairs.reduceByKey(lambda a,b: a+b)


text_id_labels = idsLabels.flatMap(lambda line: line.split("\n"))
pairs_label = text_id_labels.map(lambda s: (s.split(" ")[0],(" ").join(s.split(" ")[1:])))
jointure = pairs_label.join(counts)
#print(jointure.collect())


top = jointure.sortBy(lambda label: label[1][-1], False).collect()[:10]
top2 = []
for j in range(len(top)):
  top2.append(top[j][1])
print(top2)



```
